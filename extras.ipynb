```python


t = np.linspace(0,25,5000) ## Range da serie temporal
At = np.sin(2*np.pi*1*t) + 0.6*np.sin(2*np.pi*2*t) + 0.2*np.sin(2*np.pi*3*t) # funçao simples, só dois senos
N = len(t) # numero de pontos do role
s = np.random.normal(0, 0.25, 5000)
At = At + s


Af = np.fft.fft(At) # faz a transformada
Amag = np.abs(Af)/len(t) # usa a magnitude e normalizada

freq = np.fft.fftfreq(N,d=(t[1]-t[0])) # Sabe la deus oq rola mas ajeita as frequencias

#plota as coisa
fig, ax = plt.subplots(1,2)
fig.set_size_inches(32*0.393, 16*0.393) # diminuir na metade p 
ax[0].set_xlabel(r"$t$")
ax[0].set_ylabel(r"$A(t)$")
ax[0].set_xlim(0,2)
ax[0].plot(t,At,color = rgb_pallet[2], lw = 0.25)

ax[1].set_xlabel(r"$f$")
ax[1].set_ylabel(r"$A(f)$")
ax[1].set_xlim(0,5)
ax[1].plot(freq[0:2500],2*Amag[0:2500],color = rgb_pallet[2], lw = 1)


plt.show()


 


 
```
